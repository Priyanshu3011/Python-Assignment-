```python
Q.2.Ans:- #using for loop

su_m=0
product= 1
i=range(1,11)

for k in i :
    su_m=su_m+k
    product=product*k
      
print("sum of first 10 natural number is ",su_m,"\nproduct of first 10 natural number is",product)

#using while loop

su_m=0
product=1
i=1

while i<=10:
    su_m=su_m+i
    product=product*i
    i=i+1
    
print("sum of first 10 natural number is",su_m,"\nProduct of first natural number is ",product)
```

```python
Q.4.Ans:-

#using while and for loop

i=1

l_cube=[]
l_divisible_by_4_or_5=[]

while i <=100:
    l_cube.append(i**3)
    i=i+1
else:
    for j in l_cube:
        if j%4==0 or j%5==0:
            l_divisible_by_4_or_5.append(j)
print("cube of Each numbers\n\n",l_cube)
print("\n\n divisible by 4 or 5\n\n",l_divisible_by_4_or_5) 

# using only for loop

l=list(range(1,101))

l_cube=[]
l_divisible_by_4_or_5=[]

for i in l:
    l_cube.append(i**3)
else:
    for j in l_cube:
        if j%4==0 or j%5==0:
            l_divisible_by_4_or_5.append(j)
print(l_cube)
print(l_divisible_by_4_or_5)
```
