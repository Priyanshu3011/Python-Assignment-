```python
Q.5.Ans:-

Inheritance is a concept in object-oriented programming that allows one class to inherit properties and behavior from another class.

The class that is being inherited from is called the parent class, while the class that is inheriting is called the child class or derived class.

When a child class inherits from a parent class, it automatically gets all the properties and methods of the parent class. 

The child class can also add its own unique properties and methods, or override the ones that it inherits from the parent class.


1.Single Inheritance: In this type of inheritance, a child class inherits from a single parent class.

#For example:

class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

my_dog = Dog("Buddy")


2.Multiple Inheritance: In this type of inheritance, a child class inherits from multiple parent classes. 

#For example:

class Flyer:
    def fly(self):
        return "I can fly!"

class Swimmer:
    def swim(self):
        return "I can swim!"

class Duck(Flyer, Swimmer):
    pass

my_duck = Duck()

3.Multi-Level Inheritance: In this type of inheritance, a child class inherits from a parent class, which in turn inherits from another parent class. 

#For example:

class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model

class Car(Vehicle):
    def __init__(self, make, model, num_doors):
        super().__init__(make, model)
        self.num_doors = num_doors

class SportsCar(Car):
    def __init__(self, make, model, num_doors, has_spoiler):
        super().__init__(make, model, num_doors)
        self.has_spoiler = has_spoiler

my_car = SportsCar("Porsche", "911", 2, True)
print(my_car.make)  # Output: Porsche
```
